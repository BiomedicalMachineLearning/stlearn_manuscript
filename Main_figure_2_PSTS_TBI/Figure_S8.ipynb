```python
def read_legacy(count_matrix_file, spatial_file, image_file, gene_dict):
    data1_ex1 = st.ReadOldST(count_matrix_file=count_matrix_file,
                    spatial_file=spatial_file,
                    image_file=image_file)
    known_gene = [name for name in data1_ex1.var_names if not name.startswith('__ambiguous')]
    data1_ex1 = data1_ex1[:,known_gene]
    # If I get gene list
    with open(gene_dict, 'r') as f:
        gene_var = f.read().split('\n')
    data1_ex1.var.index = gene_var[:-1]
    data1_ex1.var_names_make_unique()
    
    return data1_ex1
```

```python
i = 3
tmp_sample = read_legacy(count_matrix_file="/home/d.pham/ST_BrainInjuryModel/JVQN_Brain_" + str(i) + "_S" + str(i) + "_stdata.tsv",
                        spatial_file="/home/d.pham/ST_BrainInjuryModel/Preprocessed/Brain" + str(i) + "_extracted_full.tsv",
                        image_file="/home/d.pham/ST_BrainInjuryModel/ClearBackground/Brain" + str(i) + ".jpg",
                        gene_dict="/home/d.pham/UQ/ST/TBI_analysis/gene_dict/ex1/gene_var_brain" + str(i) + ".txt")
```

```python
sc.pl.spatial(tmp_sample, color="Fcgr1",vmax=5,cmap="Spectral_r",size=0.5)
```

```python
sc.pl.spatial(tmp_sample, color="C1qa",vmax=5,cmap="Spectral_r",size=0.5)
```
