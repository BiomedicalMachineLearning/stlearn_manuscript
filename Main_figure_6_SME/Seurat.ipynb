```r
library(Seurat)
library(ggplot2)
library(patchwork)
library(dplyr)
```

```r
n_cluster <- 8
name <- "breast_cancer_1"
results_path <- paste0("/home/uqysun19/60days/Seurat_results/", name)
```

```r
input_path <- "/home/uqysun19/60days/10x_visium/breast_cancer_1/"
```

```r
breast_cancer_1 <- Load10X_Spatial(input_path)
```

```r
breast_cancer_1 <- SCTransform(breast_cancer_1, assay = "Spatial", verbose = FALSE)
```

```r
breast_cancer_1 <- RunPCA(breast_cancer_1, assay = "SCT", verbose = FALSE)
breast_cancer_1 <- FindNeighbors(breast_cancer_1, reduction = "pca", dims = 1:30)
breast_cancer_1 <- FindClusters(breast_cancer_1, verbose = FALSE)
breast_cancer_1 <- RunUMAP(breast_cancer_1, reduction = "pca", dims = 1:30)
```

```r
breast_cancer_1 <- FindClusters(breast_cancer_1, resolution = 0.68, verbose = FALSE)
```

```r
p <- SpatialDimPlot(breast_cancer_1, label = TRUE, label.size = 3)
```

```r
p
ggsave(paste0(results_path, "_clustering.png"), dpi=300)
```

```r
mouse_brain_coronal <- RunPCA(mouse_brain_coronal, assay = "SCT", verbose = FALSE)
mouse_brain_coronal <- FindNeighbors(mouse_brain_coronal, reduction = "pca", dims = 1:35)
# mouse_brain_coronal <- FindClusters(mouse_brain_coronal, verbose = FALSE)
mouse_brain_coronal <- RunUMAP(mouse_brain_coronal, reduction = "pca", dims = 1:35)
```

```r
mouse_brain_coronal <- FindClusters(mouse_brain_coronal, 
                                resolution = 0.93, verbose = FALSE)
```

```r
DiscretePalette(19, palette = "alphabet")
```
