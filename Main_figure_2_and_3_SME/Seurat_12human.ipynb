```r
base_path <- "/home/uqysun19/60days/Human_Brain_spatialLIBD/"
```

```r
sample = sample_list[3]
```

```r
input_path <- paste0(base_path, sample)
human_brain <- Load10X_Spatial(input_path)
gt_path <- paste0(base_path, 
                  paste0(sample, 
                         paste0("/cluster_labels_",
                                paste0(sample, ".csv"))))
ground_truth_df <- read.csv(gt_path)
n_cluster <- length(unlist(unique(ground_truth_df["ground_truth"]))) - 1
results_path <- paste0("/home/uqysun19/60days/Human_Brain_spatialLIBD/other_clustering/Seurat_results/", sample)
dir.create(results_path, showWarnings = FALSE)
human_brain <- SCTransform(human_brain, assay = "Spatial", verbose = FALSE)
human_brain <- RunPCA(human_brain, assay = "SCT", verbose = FALSE)
human_brain <- FindNeighbors(human_brain, reduction = "pca", dims = 1:30)
human_brain <- FindClusters(human_brain, verbose = FALSE)
human_brain <- RunUMAP(human_brain, reduction = "pca", dims = 1:30)



```

```r
res <- 0.1
for (i in 1:1000){
    human_brain <- FindClusters(human_brain, resolution = res, verbose = FALSE)
    n <- length(unique(human_brain@meta.data$seurat_clusters))
    if (n > n_cluster) {
        res <- res - (i/1000)*5
    }
    if (n < n_cluster) {
        res <- res + (i/1000)*5
    }
    if (n == n_cluster) {
        print(res)
        break
    }
}



```
